```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
filename = '/Users/palhares/Documents/msc/PTT-PPG-PhysioNet/physionet.org/files/pulse-transit-time-ppg/1.1.0/csv/s2_sit.csv'
column_name = 'pleth_1'
```

```python
data = pd.read_csv(filename)
time_series = data[column_name].values
```

```python
time_series = time_series[:2000]
```

```python
plt.plot(time_series)
```

```python
print(time_series)
```

```python
X = np.array(time_series).reshape(-1, 1)
```

```python
clf = IsolationForest(contamination=0.1, random_state=42)
clf.fit(X)
```

```python
y_pred = clf.predict(X)
```

```python
plt.scatter(range(len(time_series)), time_series, c=y_pred, cmap='viridis')
plt.title("Isolation Forest Anomaly Detection")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```
