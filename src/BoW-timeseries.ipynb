```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyts.approximation import SymbolicAggregateApproximation
```

```python
filename = '/Users/palhares/Documents/msc/PTT-PPG-PhysioNet/physionet.org/files/pulse-transit-time-ppg/1.1.0/csv/s2_sit.csv'
df = pd.read_csv(filename)
```

```python
column_name = 'pleth_1'
```

```python
max_rows = 5000
data = df[column_name][:max_rows]
```

```python
n_symbols = 26
```

```python
sax = SymbolicAggregateApproximation(n_bins=n_symbols, strategy='quantile')
```

```python
sax_symbols = sax.fit_transform(data.values.reshape(1, -1))
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 6))
axs[0].plot(data.index, data.values, marker='o', linestyle='-')
axs[0].set_xlabel('Amostras')
axs[0].set_ylabel('Valor')
axs[0].set_title(f'{} Primeiras Linhas de {}'.format(max_rows, column_name))
axs[1].plot(data.index, sax_symbols.flatten(), marker='o', linestyle='-')
axs[1].set_xlabel('Amostras')
axs[1].set_ylabel('SAX Letter')
axs[1].set_title(f'{max_rows} Primeiras Sequncias de Letras SAX')
plt.tight_layout()
plt.show()
```
