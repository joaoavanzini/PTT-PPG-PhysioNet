Reading the dataset - s2_sit.csv

```python
filename = '/Users/palhares/Documents/msc/PTT-PPG-PhysioNet/physionet.org/files/pulse-transit-time-ppg/1.1.0/csv/s2_sit.csv'
df = pd.read_csv(filename)
df
```

```python
df.describe()
```

```python
df.shape
```

Check if the 'time' column is present

```python
if 'time' not in df.columns:
    raise ValueError("The 'time' column does not exist in the CSV file.")
```

Transform time to seconds

```python
df['time'] = pd.to_datetime(df['time'])
df['time'] = df['time'] - df['time'][0]
df['time'] = df['time'].dt.total_seconds()
```

Fill non-numeric values with the mean

```python
df = df.apply(pd.to_numeric, errors='coerce')
df.fillna(df.mean(), inplace=True)
```

Calculate the duration of the time series

```python
start_time = df['time'].iloc[0]
end_time = df['time'].iloc[-1]
duration_seconds = end_time - start_time
```

Convert the duration to a human-readable format (e.g., minutes and seconds)

```python
duration_minutes = int(duration_seconds // 60)
remaining_seconds = int(duration_seconds % 60)
print(f"Duration of the time series: {duration_minutes} minutes and {remaining_seconds} seconds")
```

Plot the time series data - Pleth_1

```python
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['pleth_1'], label='PPG Channel 1')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('PPG Signal (Channel 1) over Time')
plt.legend()
plt.grid(True)
plt.show()
```

List of all signal columns to plot

```python
signal_columns = ['pleth_1', 'pleth_2', 'pleth_3', 'pleth_4', 'pleth_5', 'pleth_6',
                  'lc_1', 'lc_2', 'temp_1', 'temp_2', 'temp_3']
```

Create subplots for all signals

```python
plt.figure(figsize=(15, 20))
for i, column in enumerate(signal_columns, 1):
    plt.subplot(len(signal_columns), 1, i)
    plt.plot(df['time'], df[column])
    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')
    plt.title(f'Signal: {column}')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
